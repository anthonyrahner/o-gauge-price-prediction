```python
import requests
import re

url = "https://api.ebay.com/buy/browse/v1/item_summary/search"

# Fetch a small number of listings for testing
params = {
    "q": "o gauge locomotive",
    "limit": 50  
}

headers = {
    "Authorization": f'Bearer {ACCESS_TOKEN}',
    "Content-Type": "application/json"
}

response = requests.get(url, headers=headers, params=params)
print("Status code:", response.status_code)

data = response.json()

# Extract items
items = data.get("itemSummaries", [])
print(items)
```

```python
df = pd.DataFrame(items, columns=['title', 'price', 'brand', 'condition', 'has_sound', 'digital_control'])

# Sound, conventional:
#   Protosounds 1, LocoSounds, Railsounds 1-2, Lionchief

# Sound, command:
#   Protosouds 2 and 3, Railsounds 3-5, Legacy, Lionchief+ 2.0


df["price"] = df["price"].map(lambda x: float(x["value"]))



def classify_brand(title):
    if re.search(r'(?i)lionel', title):
        return "Lionel"
    elif re.search(r'(?i)mth', title):
        return "MTH"
    elif re.search(r'(?i)atlas', title):
        return "Atlas O"
    elif re.search(r'(?i)weaver', title):
        return "Weaver"
    elif re.search(r'(?i)williams', title):
        return "Williams"
    elif re.search(r'(?i)k[-]?line', title):
        return "K-Line"
    else:
        return "Other"

df["brand"] = df["title"].apply(classify_brand)

import re

def classify_sound_control(title):
    """
    Determines if a locomotive has sound and whether it's digital/command control.
    
    Returns:
        has_sound (bool)
        digital_control (bool)
    """
    title_lower = title.lower()
    
    # Sound, conventional
    sound_conventional_patterns = [
        r'protosound\s*1',
        r'locosound',
        r'railsound\s*[12]',
        r'lionchief\b(?!\+)'  # Lionchief only (not Lionchief+)
    ]
    
    # Sound, command / digital
    sound_command_patterns = [
        r'protosound\s*[23]',
        r'railsound\s*[3-5]',
        r'legacy',
        r'lionchief\+\s*2\.0?'
    ]
    
    has_sound = False
    digital_control = False
    
    # Check conventional sound systems
    for pattern in sound_conventional_patterns:
        if re.search(pattern, title_lower):
            has_sound = True
            digital_control = False  # conventional
            return has_sound, digital_control
    
    # Check command / digital sound systems
    for pattern in sound_command_patterns:
        if re.search(pattern, title_lower):
            has_sound = True
            digital_control = True  # command/digital
            return has_sound, digital_control
    
    # If no matches
    has_sound = False
    digital_control = False
    return has_sound, digital_control

df[["has_sound", "digital_control"]] = df["title"].apply(
    lambda t: pd.Series(classify_sound_control(t))
)





df
```
